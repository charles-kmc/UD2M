```python
import fnmatch
from pathlib import Path

def fast_find_webps(root: Path):
    matches = []
    for dirpath, _, filenames in os.walk(root):
        for fname in filenames:
            if fnmatch.fnmatch(fname.lower(), "*.webp"):
                matches.append(Path(dirpath) / fname)
    return matches

# Usage
webps = fast_find_webps(Path(data_dir))
```

```python
import pyodbc
import pandas as pd

def read_mdb(mdb_path: str, table_name: str):
    conn_str = (
        r'DRIVER={Microsoft Access Driver (*.mdb)};'
        rf'DBQ={mdb_path};'
    )
    conn = pyodbc.connect(conn_str)
    df = pd.read_sql(f"SELECT * FROM {table_name}", conn)
    conn.close()
    return df
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Step 1: Define the transformation (e.g., convert images to tensor and normalize)
transform = transforms.Compose([
    transforms.ToTensor(),  # Converts images to tensor
    transforms.Normalize((0.5,), (0.5,))  # Normalize with mean=0.5 and std=0.5
])

# Step 2: Load the MNIST dataset (training and test sets)
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

```

```python
# Step 3: Create DataLoader for batching
sampler = RandomSampler(train_dataset, num_samples=2000, replacement=False)
train_loader = DataLoader(train_dataset, batch_size=64, sampler = sampler)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```
